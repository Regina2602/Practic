**Задание**

- Переведите содержимое файла purchase_log.txt в словарь purchases вида: {‘1840e0b9d4’: ‘Продукты’, …}

```python
import json
i = 0
endLine=''
with open('purchase_log.txt',encoding='UTF-8') as f:
    for line in f:
        line = line.strip()
        newLine=line.replace('"user_id":','')
        newLine=newLine.replace('"category":','')
        newLine=newLine.replace(',',':')
        newLine=newLine.replace('{','')
        newLine=newLine.replace('}','')
        endLine+=newLine+','
        i += 1
        if i > 1000:
            break
        
endLine=endLine[:-1]
endLine='{'+endLine+'}'
purchases = json.loads(endLine)
print(purchases)
print(type(purchases)) #проверка типа данных
```

**Задание**
- Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

```python
with open('visit_log.csv','r',encoding='UTF-8') as f:
    with open('funnel.csv','w',encoding='UTF-8') as f_write:
        for line in f:
            line=line.strip().split(',')
            AddLine=purchases.get(line[0],None)
            if AddLine!=None:
                line.append(AddLine)
                f_write.write(line[0]+','+line[1]+','+line[2]+'\n')
            
            
            
```
